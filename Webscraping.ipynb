```python
#Example nr 1
from bs4 import BeautifulSoup
import requests, openpyxl

excel = openpyxl.Workbook()
print(excel.sheetnames)
sheet = excel.active
sheet.title = "Top Rated Movies"
print(excel.sheetnames)
sheet.append(['Rank', 'Name', 'Year of Release', 'IMDB Rating'])

html = requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
html.raise_for_status()
soup = BeautifulSoup(html.text, 'lxml')
films = soup.find('tbody', class_="lister-list").find_all('tr')

for film in films:
    name = film.find('td', class_="titleColumn").a.text
    rank = film.find('td', class_="titleColumn").get_text(strip=True).split('.')[0]
    year = film.find('span', class_="secondaryInfo").text.strip('()')
    rating = film.find('td', class_="ratingColumn imdbRating").text
    print(rank, name, year, rating)
    sheet.append([rank, name, year, rating])

excel.save('IMDB Movie Ratings')

```

```python
#Example nr 2
from bs4 import BeautifulSoup
import requests
import time

print('Write skills you are not familiar with')
unfamiliar_skill1 = input('>')
unfamiliar_skill2 = input('>')
print(f"Filtering out '{unfamiliar_skill1}' and '{unfamiliar_skill2}'")


def find_jobs():
    html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=').text
    soup = BeautifulSoup(html_text, 'lxml')
    jobs = soup.find_all('li', class_ = "clearfix job-bx wht-shd-bx")
    for job in jobs:
        job.time = job.find('span', class_="sim-posted").text
        if 'few' in job.time:
            job.info = job.header.h2.a['href']
            company_name = job.find('h3', class_="joblist-comp-name").text.replace(' ', '')
            skills = job.find('span', class_="srp-skills").text.replace(' ','')
            if unfamiliar_skill1 and unfamiliar_skill2 not in skills:
                print(f"Company name:{company_name.strip()}")
                print(f"Required skills:{skills.strip()}")
                print(f"More information here:{job.info}")
                print(' ')


if __name__ == '__main__':
    while True:
        find_jobs()
        time_wait = 10
        print(f'Waiting {time_wait} minutes... ')
        time.sleep(time_wait * 60)
```
